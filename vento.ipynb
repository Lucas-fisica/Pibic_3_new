```python
import pandas as pd
from pathlib import Path
from calendar import monthrange as mt
import re
import matplotlib.pyplot as plt
import matplotlib as mpl


mpl.rcParams['font.family'] = ['Family1', 'Serif', 'cursive']
mpl.rcParams['font.size'] = 20
```

```python
ano = 2004
pesquisa = re.compile(str(ano))
arq_c = Path()/'dados/dados'
arq_zon = [i for i in arq_c.glob('*/*.zon') if pesquisa.findall(str(i))]
arq_mer = [i for i in arq_c.glob('*/*.mer') if pesquisa.findall(str(i))]
```

```python
meses_dias = [[dias for dias in mt(year=ano, month=mes)][1] for mes in range(1, 13)]
q_dias_ano = sum(meses_dias)
colunas = [1, 3, 5, 7, 9, 11, 13]
df1 = pd.read_csv(arq_zon[1], skiprows=[0, 2], sep='\s+', dtype=float, usecols=colunas, nrows=q_dias_ano)
```

```python
fig, ax = plt.subplots(3, 1, figsize=(13, 13), sharex=True)
ax[0].plot(df1["87"], linewidth = 2, color='k', label='87Km')
ax[1].plot(df1["90"], linewidth = 2, color='r', label='90Km')
ax[2].plot(df1["93"], linewidth = 2, color='k', label='93Km')
fig.legend(loc=10, bbox_to_anchor=(0.6, 0, 0.6, 1.66))
fig.set_facecolor('#D7E2FA')
fig.set_edgecolor('#E7EBF9')
```
