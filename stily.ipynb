```python
from matplotlib import style
import matplotlib.pyplot as plt
from cmath import exp
```

```python
x = range(1, 400)
y = [exp(i/100*1j) for i in x]
```

```python
style.use('estilos/estilo2.mplstyle')
plt.plot(x, y, label='oi')
plt.ylabel('casa')
plt.xlabel('oiiiiii Texto')
plt.title('eeste é meu texto')
plt.legend()
plt.show()
```
